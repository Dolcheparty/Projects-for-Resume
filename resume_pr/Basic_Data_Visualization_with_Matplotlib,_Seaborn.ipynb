# Visualization with Matplotlib and Seaborn

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

%matplotlib inline
```


### *Part 1: Matplotlib*

### Task 1

Exponent chart

```python
x=np.linspace(0,3,100)  # Returns evenly spaced numbers in a specified interval
plt.plot (x, np.exp(x)) # indicate what values are displayed
plt.title("Exponent")   # Title of plot
plt.xlabel('x')
plt.ylabel('exp(x)')
plt.show()
```

### Task 2

Plot graphs of functions $y = x^3$ и $y = x^2$ on one graph. 


```python
plt.figure(figsize=(10,6)) # parameters in brackets determine the size of the graph
x=np.linspace(-10,10,100)
y=x**3
z=x**2
plt.plot(x,y, 'c--',label='x^3')
plt.plot(x,z, 'r:', label='x^2')
plt.title('x squared and x to the 3rd power') 
plt.xlabel('х')
plt.ylabel('y, z') 
plt.xlim((-20,20))
plt.ylim((-30,30))
plt.legend() 
plt.show()
```

### Task 3

**Scatter plot** for IRIS dataset


```python
from sklearn import datasets
data = datasets.load_iris() # loaded the iris dataset
X = data.data               # data of species (array[float])
y = data.target             # names of species (array[int])

target_names = data.target_names   # list of species' names (array[List(str)])
feature_names = data.feature_names # list of features' names (List(str))

for i in range(4):
    plt.figure(1,figsize=(10,10))
    plt.subplot(2,2,i+1)
    plt.scatter(range(150), X[:, i], c=y)
    plt.xlabel('index')
    plt.ylabel(feature_names[i])
plt.show()
```

### Task 4 

**A histogram** for the first feature of the dataset

```python
feature = 0
plt.hist(X[y==0,feature])
plt.hist(X[y==1,feature])
plt.hist(X[y==2,feature])
plt.title('Histogram of the first feature')
plt.xlabel(feature_names[0])
plt.ylabel('number of samples')
plt.legend(target_names)
```

### Task 5 

Calculate the correlation matrix between features and plot it as **a heatmap**

```python
iris = pd.DataFrame(X,y)
plt.imshow(iris.corr(), cmap='hot')
plt.colorbar() 
plt.xticks(range(4),feature_names, rotation=40, color='red')
plt.show()
```

### *Part 2: Seaborn*

### Task 6

Plot all pair interactions with marking colors in accordance with the class marks (Occupancy column)

```python
data = pd.read_csv('datatraining.txt')
data.index = pd.to_datetime(data.date)
data.head()
```

```python
sns.pairplot(data.drop('date', axis=1), hue='Occupancy')
plt.show()
```

### Task 7

**Boxplot** of all features


```python
data.drop('date', axis=1, inplace=True)
```

```python
def scale(df):
    return (df - df.mean()) / df.std()
plt.figure(figsize=(18,6))
sns.boxplot(data=scale(data.drop('Occupancy', axis=1)), orient='h')
plt.show()
```

### Task 8

Violinplot of all features



```python
plt.figure(figsize=(12,6))
sns.violinplot(data=scale(data.drop('Occupancy', axis=1)), orient='h')
plt.show()
```

### Task 8

**Heatmap** of correlation matrix

```python
corr = data.drop('Occupancy', axis=1).corr()
sns.heatmap(corr, cmap=sns.color_palette("RdBu_r", 1000), vmin=-1, center=0, annot=True)
plt.show()
```

### Task 9 

Boxplot of all features presence by Occupancy. 

```python
df = pd.read_csv('datatraining.txt')
df.index = pd.to_datetime(df.date)
df.insert(0,'StrOccupancy', df.Occupancy.values) 
df.replace({'StrOccupancy':{0:'No',1:'Yes'}})
```

```python
for col in df.columns[2:-1]:
  sns.boxplot(df.StrOccupancy,df[col], data = df)
  plt.show()
```
